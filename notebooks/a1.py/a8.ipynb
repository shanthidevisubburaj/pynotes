```python
text = "madam"
print("Palindrome?", text == text[::-1])
```

```python
num = 153
digits = str(num)
total = sum(int(d)**3 for d in digits)
print("Armstrong?", total == num)
```

```python
text = "Hello Python"
vowels = "aeiouAEIOU"
count = sum(1 for ch in text if ch in vowels)
print("Vowel count:", count)
```
